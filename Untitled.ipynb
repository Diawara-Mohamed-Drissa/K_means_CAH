```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from scipy.cluster.hierarchy import fcluster


df = pd.read_csv('Mall_Customers.csv', encoding='utf-8')
print(df.head())  
```

##  1. Statistiques descriptives (Age, Income, Spending Score

```python
age_moyen = np.mean(df["Age"])
age_max = np.max(df["Age"])
age_min = np.min(df["Age"])
age_var = np.var(df["Age"])
age_ecart = np.std(df["Age"])
age_mediane = np.median(df["Age"])
mode_age = df["Age"].mode()

print("Âge moyen       :", age_moyen)
print("Âge maximum     :", age_max)
print("Âge minimum     :", age_min)
print("Variance de l'âge :", age_var)
print("Écart-type de l'âge :", age_ecart)
print("Médiane de l'âge :", age_mediane)
print("Mode de l'âge    :", list(mode_age))


#68% de la population
borne_inf = age_moyen - age_ecart
borne_inf
borne_sup = age_moyen + age_ecart
borne_sup

print(f"68% de la population sont agés entre {borne_inf} et {borne_sup}")


#95% de la population
borne_inf = age_moyen - 2 * age_ecart
borne_inf
borne_sup = age_moyen + 2 * age_ecart
borne_sup

print(f"95% de la population sont agés entre {borne_inf} et {borne_sup}")


#99% de la population
borne_inf = age_moyen - 3 * age_ecart
borne_inf
borne_sup = age_moyen + 3 * age_ecart
borne_sup

print(f"99% de la population sont agés entre {borne_inf} et {borne_sup}")


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df["Age"], bins=15, edgecolor='black')
plt.title("Distribution de l'âge")
plt.xlabel("Âge")
plt.ylabel("Nombre de clients")


```

####  3. Normalisez les données

```python
for col in ["Age", "Annual Income (k$)", "Spending Score (1-100)"]:
    moyenne = df[col].mean()
    ecart_type = df[col].std()
    df[col + "_standardisee"] = (df[col] - moyenne) / ecart_type

# Afficher les 5 premières lignes
print(df[["Age_standardisee", "Annual Income (k$)_standardisee", "Spending Score (1-100)_standardisee"]].head())
```

###  Matrice de variance-covariance et ACP

####  1. Calculez la matrice de variance-covariance des variables numériques

#### Manuel

```python
colonnes_std = [
    "Age_standardisee",
    "Annual Income (k$)_standardisee",
    "Spending Score (1-100)_standardisee"
]

matrice_standard = df[colonnes_std].values
colonnes_standard = [
    "Age_standardisee",
    "Annual Income (k$)_standardisee",
    "Spending Score (1-100)_standardisee"
]

cov_matrix_standard = df[colonnes_standard].cov()
print("\n Matrice de variance-covariance (standardisée) :\n")
print(cov_matrix_standard)

```

####  2. Interprétez les variances et covariances

#### 
1. Âge  ,Revenu : -0.012

 Corrélation quasiment nulle
 L’âge n’a aucune relation avec le revenu.

2. Âge,  Spending Score : -0.327

  Corrélation négative modérée
  Plus l’âge est élevé, plus le spending score est faible.
  Les jeunes dépensent plus, les plus âgés moins.

3. Revenu ,  Spending Score : +0.009

  Corrélation nulle
  Le revenu n’a aucun lien avec le spending score.
  Avoir un revenu élevé ne signifie pas dépenser plus dans ce dataset.

###  3. Appliquez l’ACP sur les variables normalisées

```python
valeurs_propres, vecteurs_propres = np.linalg.eig(cov_matrix_standard)
index = valeurs_propres.argsort()[::-1]  # tri du plus grand au plus petit
valeurs_2 = valeurs_propres[index][:2]
vecteurs_2 = vecteurs_propres[:, index][:, :2]
print("Les 2 plus grandes valeurs propres :")
print(valeurs_2)

print("\nLes 2 vecteurs propres associés :")
print(vecteurs_2)
```

#### 4. Déterminez le nombre de composantes à retenir (variance expliquée)

```python
nb = len(valeurs_propres)
print(f"Le nombre de composantes est : {nb}")
variance_expliquee = valeurs_propres / valeurs_propres.sum()

# 4. Variance expliquée cumulée
variance_cumulee = np.cumsum(variance_expliquee)

print("Valeurs propres : ", valeurs_propres)
print("Variance expliquée : ", variance_expliquee)
print("Variance expliquée cumulée : ", variance_cumulee)
```

####  5. Visualisez le cercle des corrélations

```python
colonnes = ["Age", "Income", "Spending Score"]

# Coordonnées des variables dans le cercle
corr = vecteurs_propres * np.sqrt(valeurs_propres)

x = corr[:, 0]
y = corr[:, 1]

plt.figure(figsize=(6,6))

# Cercle unité
theta = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(theta), np.sin(theta))

# Axes
plt.axhline(0)
plt.axvline(0)

# Flèches + noms
for i in range(len(colonnes)):
    plt.arrow(0, 0, x[i], y[i], head_width=0.03)
    plt.text(x[i]*1.12, y[i]*1.12, colonnes[i])

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Cercle des corrélations")
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.gca().set_aspect('equal', 'box')
plt.show()
```

1. Age

Flèche longue vers le bas → fortement lié à PC1

Peu lié à PC2
    PC1 représente essentiellement l’âge
    Plus PC1 est élevé → personnes plus âgées

2. Spending Score

Flèche longue vers le haut-gauche → opposé à l’âge

Bien représenté dans le plan
    Les jeunes dépensent plus, les plus âgés dépensent moins
    Corrélation clairement négative avec l’âge

3. Income (Revenu)

Flèche très courte → faible représentation
    Le revenu n’est pas bien expliqué par PC1–PC2
    Peu ou pas corrélé aux deux axes

###  Classification Hiérarchique Ascendante (CAH)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# On utilise tes données standardisées
X = matrice_standard   # si tu as déjà créé cette variable; sinon on la calcule

plt.figure(figsize=(15, 6))

# Complete Linkage 
plt.subplot(1, 3, 1)
Z_complete = linkage(X, method='complete')
dendrogram(Z_complete, no_labels=True)
plt.title("Complete Linkage")

# Average Linkage
plt.subplot(1, 3, 2)
Z_average = linkage(X, method='average')
dendrogram(Z_average, no_labels=True)
plt.title("Average Linkage")

# Ward Linkage
plt.subplot(1, 3, 3)
Z_ward = linkage(X, method='ward')
dendrogram(Z_ward, no_labels=True)
plt.title("Ward Linkage")

plt.tight_layout()
plt.show()
```

###  2. Choisissez le meilleur critère et coupez le dendrogramme

####
Nous avons retenu la méthode de Ward car elle minimise la variance intra-cluster à chaque fusion, ce qui produit des groupes plus homogènes et équilibrés. Parmi les trois méthodes testées (Complete, Average, Ward), le dendrogramme obtenu avec Ward présente la plus grande hauteur de fusion finale, révélant une meilleure séparation entre les clusters. De plus, la méthode de Ward est particulièrement adaptée aux données quantitatives et donne d'excellents résultats après réduction de dimension (PCA), ce qui correspond parfaitement à notre jeu de données.

####  5. Appliquez la CAH sur les composantes principales : observe-t-on une meilleure séparation?

```python
# 1. CAH avec Ward sur les composantes principales
Z_pc = linkage(projection, method='ward')

# 2. Dendrogramme
plt.figure(figsize=(12, 6))
dendrogram(Z_pc)
plt.title("Dendrogramme CAH (sur PC1 et PC2)")
plt.show()

# 3. Découpage en 3 clusters (ou 4 si tu veux suivre la logique du dendrogramme)
clusters_pc = fcluster(Z_pc, t=4, criterion='maxclust')

clusters_pc
```

```python
# 1. CAH avec Ward sur les composantes principales
Z_pc = linkage(projection, method='ward')

# 2. Dendrogramme
plt.figure(figsize=(12, 6))
dendrogram(Z_pc)
plt.title("Dendrogramme CAH (sur PC1 et PC2)")
plt.show()

# 3. Découpage en 4 clusters (ou 4 si tu veux suivre la logique du dendrogramme)
clusters_pc = fcluster(Z_pc, t=4, criterion='maxclust')

clusters_pc
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(projection[:,0], projection[:,1], c=clusters_pc, cmap="viridis")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters CAH sur les composantes principales")
plt.show()
```

####
Oui.
En appliquant la CAH sur les deux composantes principales, on observe une meilleure séparation visuelle des groupes :

Les points sont plus regroupés à l’intérieur de chaque cluster.

Les clusters sont plus éloignés les uns des autres.

Les frontières entre groupes sont plus nettes qu’avec les données brutes.

###  Évaluation et validation

#### 1. Calculez le coefficient de silhouette pour chaque méthode

```python
# On prend les CAH sur les données PCA (projection = ta matrice 200x2)
# CHOISIS le nombre de clusters, par ex. 4 (d'après ton graphe)
k = 4
clusters_ward = fcluster(Z_ward, t=k, criterion='maxclust')
clusters_complete = fcluster(Z_complete, t=k, criterion='maxclust')
clusters_average = fcluster(Z_average, t=k, criterion='maxclust')

ward_score = silhouette_score(projection, clusters_ward)
complete_score = silhouette_score(projection, clusters_complete)
avg_score = silhouette_score(projection, clusters_average)

print("Silhouette - Ward     :", ward_score)
print("Silhouette - Complete :", complete_score)
print("Silhouette - Average  :", avg_score)

```

#### Dans la plupart des cas ward est meilleur mais ce ne n'est pas le cas ici car on n'a pas de clusters ronds mais plutot etirés et average mieux avec les cluster etirés

####  2. Calculez l’indice de Davies-Bouldin

```python
db_ward = davies_bouldin_score(projection, clusters_ward)
db_complete = davies_bouldin_score(projection, clusters_complete)
db_average = davies_bouldin_score(projection, clusters_average)

print("Davies-Bouldin - Ward     :", db_ward)
print("Davies-Bouldin - Complete :", db_complete)
print("Davies-Bouldin - Average  :", db_average)
print("Avec Davies-Bouldin plus le score est faible plus la methode est meilleure")
```
