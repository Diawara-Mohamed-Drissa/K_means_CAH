###  Matrice de variance-covariance et ACP

####  1. Calculez la matrice de variance-covariance des variables numériques

#### Manuel

```python
colonnes_std = [
    "Age_standardisee",
    "Annual Income (k$)_standardisee",
    "Spending Score (1-100)_standardisee"
]

matrice_standard = df[colonnes_std].values
colonnes_standard = [
    "Age_standardisee",
    "Annual Income (k$)_standardisee",
    "Spending Score (1-100)_standardisee"
]

cov_matrix_standard = df[colonnes_standard].cov()
print("\n Matrice de variance-covariance (standardisée) :\n")
print(cov_matrix_standard)

```

###  3. Appliquez l’ACP sur les variables normalisées

```python
valeurs_propres, vecteurs_propres = np.linalg.eig(cov_matrix_standard)
index = valeurs_propres.argsort()[::-1]  # tri du plus grand au plus petit
valeurs_2 = valeurs_propres[index][:2]
vecteurs_2 = vecteurs_propres[:, index][:, :2]
print("Les 2 plus grandes valeurs propres :")
print(valeurs_2)

print("\nLes 2 vecteurs propres associés :")
print(vecteurs_2)
```

#### 4. Déterminez le nombre de composantes à retenir (variance expliquée)

```python
nb = len(valeurs_propres)
print(f"Le nombre de composantes est : {nb}")
variance_expliquee = valeurs_propres / valeurs_propres.sum()

# 4. Variance expliquée cumulée
variance_cumulee = np.cumsum(variance_expliquee)

print("Valeurs propres : ", valeurs_propres)
print("Variance expliquée : ", variance_expliquee)
print("Variance expliquée cumulée : ", variance_cumulee)
```

####  5. Visualisez le cercle des corrélations

```python
colonnes = ["Age", "Income", "Spending Score"]

# Coordonnées des variables dans le cercle
corr = vecteurs_propres * np.sqrt(valeurs_propres)

x = corr[:, 0]
y = corr[:, 1]

plt.figure(figsize=(6,6))

# Cercle unité
theta = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(theta), np.sin(theta))

# Axes
plt.axhline(0)
plt.axvline(0)

# Flèches + noms
for i in range(len(colonnes)):
    plt.arrow(0, 0, x[i], y[i], head_width=0.03)
    plt.text(x[i]*1.12, y[i]*1.12, colonnes[i])

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Cercle des corrélations")
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.gca().set_aspect('equal', 'box')
plt.show()
```

### K-Means

####  1. Déterminez k optimal avec : méthode du coude + silhouette

```python
x = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
x_scaled = StandardScaler().fit_transform(x)

#Methode du coude + silhouette 
inertias = []
silhouettes = []
rang = range(2,11)

for k in rang:
    kmeans = KMeans(n_clusters = k, random_state = 0, n_init='auto')
    labels = kmeans.fit_predict(x_scaled)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(x_scaled, labels))
    
plt.figure(figsize=(8,6))
plt.plot(rang, inertias, marker='o')
plt.title("Methode du coude")
plt.xlabel("k")
plt.ylabel("Inertie")
plt.grid(True)
plt.show()
    
plt.figure(figsize=(8,6))
plt.plot(rang, silhouettes, marker='o')
plt.title("Coefficient de silhouette")
plt.xlabel("k")
plt.ylabel("Silhouette")
plt.grid(True)
plt.show()




```

#### 2. Appliquez K-Means avec k optimal sur les données originales

```python
k_op = rang[np.argmax(silhouettes)]
print("k optimal :",k_op )
```

```python
kmeans_og = KMeans(n_clusters = k_op, random_state = 0, n_init = 'auto')
df["Cluster_original"] = kmeans_og.fit_predict(x_scaled)
df
```

#### 3. Appliquez K-Means sur les composantes principales retenues

```python
# PARTIE PCA

pca = PCA()                     # <-- créer l'objet PCA
x_pca = pca.fit_transform(x_scaled)

# Nombre de composantes nécessaires pour 90% variance
var = np.cumsum(pca.explained_variance_ratio_)
comp = np.argmax(var >= 0.90) + 1
print("Nombre de composantes PCA retenues :", comp)

# Réduction avec le nombre optimal de composantes
x_pca_red = PCA(n_components=comp).fit_transform(x_scaled)

# KMeans sur la PCA
kmeans_pca = KMeans(n_clusters=k_op, random_state=0, n_init='auto')
df["Cluster_pca"] = kmeans_pca.fit_predict(x_pca_red)

df.head()
```

#### 4. Visualisez en scatter plot : Income vs Spending Score (coloré par cluster)

```python
#Scatter plot : Income vs Spending Score 
plt.figure(figsize=(8,6))
sns.scatterplot(
    data = df,
    x = "Annual Income (k$)",
    y = "Spending Score (1-100)",
    hue = "Cluster_original",
    palette = "tab10",
    s = 60
)
plt.title("Income vs Spending Score")
plt.grid(True)
plt.show()
```

```python
# KMeans sur PC1-PC2 avec 4 clusters
kmeans_pc2 = KMeans(n_clusters=4, random_state=0, n_init='auto')

clusters_pc2 = kmeans_pc2.fit_predict(x_pca[:, :2])   # ou projection si tu l'as déjà

df["Cluster_PC2"] = clusters_pc2

plt.figure(figsize=(8,6))
plt.scatter(
    x_pca[:,0], 
    x_pca[:,1], 
    c=df["Cluster_PC2"], 
    cmap="viridis", 
    s=60
)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clustering K-Means (4 clusters) sur PC1 et PC2")
plt.grid(True)
plt.show()
```

#### 6. Comparez les résultats K-Means sur données originales vs sur composantes principales

```python
#Comparaison 
print(pd.crosstab(df['Cluster_original'], df['Cluster_pca']))
```

```python
silhouette_orig = silhouette_score(x_scaled, df['Cluster_original'])
silhouette_pca = silhouette_score(x_pca_red, df['Cluster_pca'])

print(f"Coeff de silhouette: \n -original : {silhouette_orig:.4f} \n -pca : {silhouette_pca:.4f}")



#Indice de Davies-Bouldin
db_orig = davies_bouldin_score(x_scaled, df['Cluster_original'])
db_pca = davies_bouldin_score(x_pca_red, df['Cluster_pca'])

print("Indice de Davies-Bouldin :")
print(f"Données originales : {db_orig:.4f} et Composantes principales (PCA) : {db_pca:.4f}")

#Meilleur chemin
if silhouette_pca > silhouette_orig and db_pca < db_orig:
    print("K-Means sur PCA produit les meilleurs clusters (meilleure silhouette et DB plus faible).")
elif silhouette_orig > silhouette_pca and db_orig < db_pca:
    print("K-Means sur données originales produit les meilleurs clusters.")
else:
    print("Résultats mitigés")

#ACP
improve_silhouette = silhouette_pca > silhouette_orig
improve_db = db_pca < db_orig

if improve_silhouette and improve_db:
    print("L'utilisation de l'ACP améliore la qualité du clustering.")
else:
    print("L'utilisation de l'ACP n'améliore pas significativement la qualité du clustering.")
```

###  Classification Hiérarchique Ascendante (CAH)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# On utilise tes données standardisées
X = matrice_standard   # si tu as déjà créé cette variable; sinon on la calcule

plt.figure(figsize=(15, 6))

# Complete Linkage 
plt.subplot(1, 3, 1)
Z_complete = linkage(X, method='complete')
dendrogram(Z_complete, no_labels=True)
plt.title("Complete Linkage")

# Average Linkage
plt.subplot(1, 3, 2)
Z_average = linkage(X, method='average')
dendrogram(Z_average, no_labels=True)
plt.title("Average Linkage")

# Ward Linkage
plt.subplot(1, 3, 3)
Z_ward = linkage(X, method='ward')
dendrogram(Z_ward, no_labels=True)
plt.title("Ward Linkage")

plt.tight_layout()
plt.show()
```

###  2. Choisissez le meilleur critère et coupez le dendrogramme

####
Nous avons retenu la méthode de Ward car elle minimise la variance intra-cluster à chaque fusion, ce qui produit des groupes plus homogènes et équilibrés. Parmi les trois méthodes testées (Complete, Average, Ward), le dendrogramme obtenu avec Ward présente la plus grande hauteur de fusion finale, révélant une meilleure séparation entre les clusters. De plus, la méthode de Ward est particulièrement adaptée aux données quantitatives et donne d'excellents résultats après réduction de dimension (PCA), ce qui correspond parfaitement à notre jeu de données.

####  3. Matrice de confusion : comparez K-Means vs CAH

```python
#1) Nombre de clusters
k = 4

# --- 2) CAH avec Ward sur PC1–PC2 ---
clusters_ward = fcluster(Z_ward, t=k, criterion='maxclust')
df["Cluster_cah"] = clusters_ward

# --- 3) Matrice de confusion entre K-Means PCA et CAH PCA ---
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(df["Cluster_pca"], df["Cluster_cah"])

# 4) Affichage 
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
            xticklabels=[f"CAH {i}" for i in range(1, k+1)],
            yticklabels=[f"KMeans {i}" for i in range(1, k+1)])
plt.xlabel("Clusters CAH (Ward)")
plt.ylabel("Clusters K-Means (PCA)")
plt.title("Matrice de confusion : K-Means PCA vs CAH PCA")
plt.show()
```

####  4. Les deux méthodes identifient-elles les mêmes groupes?

####
Les clusters obtenus par K-Means et par la CAH ne sont pas identiques.
Les deux méthodes retrouvent des profils similaires (jeunes fort dépensiers, clients âgés à faible score, groupe intermédiaire),
mais la partition exacte varie.
Cela s’explique par les différences entre les deux algorithmes :
K-Means produit des clusters sphériques, tandis que la CAH utilise une logique hiérarchique.
